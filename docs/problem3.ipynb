```python
import altair as alt
import pandas as pd

tanzaniaData = pd.read_csv("./data/tanzaniaFertility.csv")

tanzaniaData
```

```python
tanzaniaData['year'] = pd.to_datetime(tanzaniaData['year'], format='%Y')

tanzaniaDataLong = tanzaniaData.melt('year', var_name='indicator', value_name='value')

readable_labels = {
    'fertilityRate': 'Fertility Rate',
    'contraceptionRate': 'Contraception Usage',
    'unmetNeedPercent': 'Unmet Family Planning'
}

tanzaniaDataLong['indicator'] = tanzaniaDataLong['indicator'].map(readable_labels)

chart = alt.Chart(tanzaniaDataLong).mark_line(point=True).encode(
    x=alt.X('year:T', axis=alt.Axis(title='Year', format="%Y")),
    y=alt.Y('value:Q', axis=alt.Axis(title='Rate (%)')),
    color=alt.Color('indicator:N', legend=alt.Legend(title="")),
    strokeDash=alt.StrokeDash('indicator:N', legend=None),
    tooltip=['year:T', 'value:Q']
).properties(
    title="Contraception Rate and Unmet Family Planning Doesn't Mitigate Fertitlity",
    width=500,
    height=400
)

chart
```

c. Through the visualization displayed you are easily able to see the story in Tanzania which is that even though there was consistentcy among the use of Contraception Methods and Unmet Need for Family Planning Percent. Fertility Rate skyrocketed suggesting these measure have little effect on the fertility rate. I dealt with the dates stretching over two years by going to the latest year as that would be the year with which the data was finished collecting.
