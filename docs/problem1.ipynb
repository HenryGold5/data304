```python
import altair as alt
import pandas as pd


twinData = pd.read_csv("https://calvin-data304.netlify.app/data/twins-genetics-long.csv")

twinData.head()
```

Comparing Twins

```python
mean_genetic_share = twinData.groupby(['pair', 'id', 'region'])['genetic share'].mean().reset_index()
pivot_df = mean_genetic_share.pivot_table(index=['pair', 'region'], columns='id', values='genetic share')
pivot_df['difference'] = pivot_df['A'] - pivot_df['B']
diff_df = pivot_df.reset_index()
diff_df.columns = diff_df.columns.get_level_values(0)

chart = alt.Chart(diff_df).mark_bar().encode(
    x=alt.X('region:N', title='Region'),
    y=alt.Y('difference:Q', title='Difference Genetic Share'),
    color=alt.condition(
        alt.datum.difference > 0,
        alt.value("steelblue"),
        alt.value("red")
    ),
    tooltip=['pair:N', 'region:N', 'difference:Q']
).properties(
    width=alt.Step(20),
    height=150
).facet(
    facet=alt.Facet('pair:N', title="Twin"),
    columns=3
).properties(
    title=alt.TitleParams("Twins have Similar Genetic Shares")
)

chart.display()
```

Comparing Kits

```python
chart = alt.Chart(twinData).mark_bar().encode(
    x=alt.X('region:N', title='Region'),
    y=alt.Y('genetic share:Q', title='Genetic Share'),
    color=alt.Color('kit:N', title='Test Kit'),
    xOffset='kit:N',
    tooltip=['twin', 'region', 'kit', 'genetic share']
).properties(
    width=200,
    height=100
).facet(
    facet=alt.Facet('twin:N', title='Twin ID'),
    columns=2 
).properties(
    title=alt.TitleParams("Ancestry Has Greater Variance then Other Test Kits")
)

chart
```
