```python
import altair as alt
import pandas as pd
import pyreadstat

pewCenterData, meta = pyreadstat.read_sav("./data/pewResearchRaw.sav")

pewCenterDataParties2016 = pewCenterData[["VOTEPARTY_W16", "WINLOSE_W16", "MATTERS_W16", "THOUGHT_W16"]]

pewCenterDataViewsOnIssue = pewCenterData[["VOTEPARTY_W16", "PTYISSUE_A_W16","PTYISSUE_B_W16","PTYISSUE_C_W16","PTYISSUE_D_W16","PTYISSUE_E_W16","PTYISSUE_F_W16","PTYISSUE_G_W16","PTYISSUE_H_W16","PTYISSUE_I_W16"]]
```

```python
pewCenterDataParties2016.head()
```

```python
pewCenterDataViewsOnIssue['PoliticalParty'] = pewCenterDataViewsOnIssue['VOTEPARTY_W16'].map({
    1: 'Republican',
    2: 'Republican',
    3: 'Democrat',
    4: 'Democrat'
})

pewCenterDataViewsOnIssue['SupportLevel'] = pewCenterDataViewsOnIssue['VOTEPARTY_W16'].map({
    1: 'Definitely Republican',
    2: 'Somewhat Republican',
    3: 'Somewhat Democrat',
    4: 'Definitely Democrat'
})

pewCenterDataViewsOnIssue.head()
```

```python
column_name_mapping = {
    'PTYISSUE_A_W16': 'HealthCarePolicy',
    'PTYISSUE_B_W16': 'IllegalImmigrationPolicy',
    'PTYISSUE_C_W16': 'ClimateChangePolicy',
    'PTYISSUE_D_W16': 'ISISPolicy',
    'PTYISSUE_E_W16': 'AbortionPolicy',
    'PTYISSUE_F_W16': 'GunPolicy',
    'PTYISSUE_G_W16': 'EconomicPolicy',
    'PTYISSUE_H_W16': 'WealthGapPolicy',
    'PTYISSUE_I_W16': 'BudgetDeficitPolicy' 
}

pewCenterDataViewsOnIssue.rename(columns=column_name_mapping, inplace=True)

response_option_mapping = {
    1: 'Strongly agree with the Republican Party',
    2: 'Somewhat agree with the Republican Party',
    3: 'Somewhat agree with the Democratic Party',
    4: 'Strongly agree with the Democratic Party',
    5: 'Don’t agree with either party'
}

for column in column_name_mapping.values():
    pewCenterDataViewsOnIssue[column] = pewCenterDataViewsOnIssue[column].map(response_option_mapping)

pewCenterDataViewsOnIssue.head()
```

```python
pewCenterDataViewsOnIssue.drop('VOTEPARTY_W16', axis=1, inplace=True)

pewCenterDataViewsOnIssueLong = pd.melt(pewCenterDataViewsOnIssue, id_vars=['PoliticalParty', 'SupportLevel'], 
                  var_name='PolicyArea', value_name='AgreementLevel')

pewCenterDataViewsOnIssueLong.head()
```

```python
pewCenterDataViewsOnIssueLong['AgreementLevel'] = pewCenterDataViewsOnIssueLong['AgreementLevel'].fillna('')
df_clean = pewCenterDataViewsOnIssueLong.dropna(subset=['PoliticalParty'])

def agreement_with_own_party(row):
    if row['PoliticalParty'] == 'Republican' and 'Republican Party' in row['AgreementLevel']:
        return 'Agree'
    elif row['PoliticalParty'] == 'Democrat' and 'Democratic Party' in row['AgreementLevel']:
        return 'Agree'
    else:
        return 'Disagree'
    
df_clean['AgreesWithOwnParty'] = df_clean.apply(agreement_with_own_party, axis=1)

policy_agreement_counts = df_clean.groupby(['PoliticalParty','PolicyArea', 'AgreesWithOwnParty']).size().unstack(fill_value=0)
policy_agreement_counts['AgreePercentage'] = policy_agreement_counts['Agree'] / (policy_agreement_counts['Agree'] + policy_agreement_counts['Disagree']) * 100

policy_agreement_percentages = policy_agreement_counts.reset_index()[['PoliticalParty','PolicyArea', 'AgreePercentage']]

print(policy_agreement_percentages)
```

```python
pivot_df = policy_agreement_percentages.pivot(index='PolicyArea', columns='PoliticalParty', values='AgreePercentage').reset_index()

pivot_df.columns = ['PolicyArea', 'Democrat', 'Republican']

pivot_df['Difference'] = abs(pivot_df['Democrat'] - pivot_df['Republican'])

pivot_df['DifferenceText'] = pivot_df.apply(lambda row: f"{abs(row['Difference']):.1f}% (R>D)" 
                                             if row['Republican'] > row['Democrat'] 
                                             else f"{abs(row['Difference']):.1f}% (D>R)", axis=1)

base = alt.Chart(pivot_df).encode(
    y=alt.Y('PolicyArea:N', axis=alt.Axis(title='')),
)

republican_points = base.mark_circle(color='red', size = 30).encode(
    x=alt.X('Republican:Q', axis=alt.Axis(title='Agrees with Party Percentage')),
    tooltip=['PolicyArea:N', alt.Tooltip('Republican:Q', format='.2f')]
)

democrat_points = base.mark_circle(color='blue', size=30).encode(
    x='Democrat:Q',
    tooltip=['PolicyArea:N', alt.Tooltip('Democrat:Q', format='.2f')]
)

lines = base.mark_line().encode(
    x='Republican:Q',
    x2='Democrat:Q',
    tooltip=['PolicyArea:N', alt.Tooltip('Difference:Q', format='.2f')]
)

chart = alt.layer(republican_points, democrat_points, lines).properties(
    width=300,
    height=300,
    title='Disagreement within Parties on Public Policy'
)

difference_text_chart = alt.Chart(pivot_df).mark_text().encode(
    y=alt.Y('PolicyArea:N', axis=alt.Axis(title='', labels=False), sort=None),
    text='DifferenceText:N'
).properties(
    width=100,  # Adjust width as necessary
    height=300,  # Adjust height to match the main chart if displaying side-by-side
    title=''
)


chart | difference_text_chart
```

For this graphic I wanted to use Pew Research Center Data to see if I could wrangle it and then make a cool graphic with it because they always have fascinating data sets on the american public and with the election coming up I thought it would be quite pertient. The data wrangling was quite a struggle but I eventually got their. For my graphic I've seen a lot of times especially when dealing with political parties a dumbbell graphic that are really cool because they can show easily where party lines differ. I had thought of other alternatives such as stacked or grouped bar charts as well as a boxplot to convey this data but I wanted to do a more challenging graphic and see if I could recreate it in altair. I had done this kind of graphic for work in ggplot so it was fun to be able to implement it again in vegalite. For the color I choose red for the republican party as well as blue for the democratic party and then didn't use a legend for the colors because these are commonly respected in political data and easily knowable. I also added tooltips to the points to show the percentages and then the party to be clear on the data being presented. For the lines tooltip I decided to show the absolute difference so viewers could easily see what the actual difference between the two parties were on each policy issue. The title I wanted to get across the main message of this graph which was that even though people align with a certain party they don't accept all the views on every policy issue when it comes to the wealth gap and budget deficit policy both parties had less then 50% of their constinuents agree with their parties stance on the issue. I then concated a text box in altair that used the difference so the user could easily see the differences between the Republican and Democratic Party.
