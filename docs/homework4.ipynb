```python
import altair as alt
import pandas as pd

data = [{"Date":"Q1-2017","Completion Rate":0.91,"Response Rate":0.023},
        {"Date":"Q2-2017","Completion Rate":0.93,"Response Rate":0.018},
        {"Date":"Q3-2017","Completion Rate":0.91,"Response Rate":0.028},
        {"Date":"Q4-2017","Completion Rate":0.89,"Response Rate":0.023},
        {"Date":"Q1-2018","Completion Rate":0.84,"Response Rate":0.034},
        {"Date":"Q2-2018","Completion Rate":0.88,"Response Rate":0.027},
        {"Date":"Q3-2018","Completion Rate":0.91,"Response Rate":0.026},
        {"Date":"Q4-2018","Completion Rate":0.87,"Response Rate":0.039},
        {"Date":"Q1-2019","Completion Rate":0.83,"Response Rate":0.028}] 

emailData = pd.DataFrame(data)

base = alt.Chart(emailData).encode(
    alt.X(field = "Date", type = "ordinal", title = "",
          sort=['Q1-2017', 'Q2-2017', 'Q3-2017', 'Q4-2017', 'Q1-2018', 'Q2-2018', 'Q3-2018', 'Q4-2018', 'Q1-2019'])
).properties(width = 300, height = 200)

completionLine = base.mark_line().encode(
    alt.Y(field = "Completion Rate", type = "quantitative", axis = alt.Axis(format='.0%'), scale=alt.Scale(domain=[0.8,1])),
    color = alt.value('#3a73bd'))
completionPoint = base.mark_point().encode(
    alt.Y(field = "Completion Rate", type = "quantitative", axis = alt.Axis(format='.0%'), scale=alt.Scale(domain=[0.8,1])),
    fill = alt.value('#3a73bd'),
    color = alt.value('#3a73bd')
)

completionChart = completionLine + completionPoint

responseLine = base.mark_line().encode(
    alt.Y(field = "Response Rate", type = "quantitative", axis = alt.Axis(format='.1%'), scale=alt.Scale(domain=[0,0.04])),
    color = alt.value('#fd7600')).properties(
        title = alt.Title("While Email Responses Rise, Completion Rates Fall", align='center')
    )
responsePoint = base.mark_point().encode(
    alt.Y(field = "Response Rate", type = "quantitative", axis = alt.Axis(format='.1%'), scale=alt.Scale(domain=[0,0.04])),
    fill = alt.value('#fd7600'),
    color = alt.value('#fd7600'))

responseChart = responseLine + responsePoint
responseChart | completionChart
```
